### Проверка согласованности матриц

```python
def vector_function(matrix):
    matrix = np.array(matrix, dtype=float)
    column_sums = np.sum(matrix, axis=0)
    matrix_new = matrix / column_sums
    row_sums = np.sum(matrix_new, axis=1)
    vector = row_sums / np.sum(row_sums)
    return vector
```

```python
def cheking_consistency(target_kriterias):
    for i in target_kriterias: 
        vector = vector_function(i)
        lamda_max = np.mean(np.dot(i, vector) / vector)
        IS = (lamda_max - len(vector)) / (len(vector) - 1)
        OS = IS / 1.12
        if OS < 0.1:
            print(f"OS равно {round(OS, 4)}, согласованность приемлема")
        else:
            print(f"OS равно {round(OS, 4)}, согласованность НЕ приемлема")
```

```python
cheking_consistency(target_kriterias)
```
