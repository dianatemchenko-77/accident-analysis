### Преобразование типов данных

```python
#приводим столбец crash_date к виду год/месяц/день, т.е. добавляем разделители
c = 0
for i in df['crash_date']:
    df.loc[c, 'crash_date'] = str(i)[0:4] +'/'+ str(i)[4:6] + '/' + str(i)[6:]
    c+=1
```

```python
#меняем тип столбца на datatime64
df['crash_date'] = pd.to_datetime(df['crash_date'], format='%Y/%m/%d')
df['crash_date'] = pd.to_datetime(df['crash_date']).dt.normalize()
df
```

```python
#приводим столбец crash_time к виду времени, однако, тип столбца остается object
df['crash_time'] = pd.to_datetime(df['crash_time'], format='%H:%M:%S').dt.time
df
```

```python
df[['participants_amount']] = df[['participants_amount']].apply(pd.to_numeric) #меняем тип на int
```

```python
df.info()
```

### Поиск дубликатов

```python
df.loc[df.duplicated(keep=False)] #находим дубликаты
```

```python
df = df.drop_duplicates().reset_index(drop=True) #удаляем дубликаты
df
```

### Добавление столбцов.

```python
#создаем столбец full_datatime(=datatime) путем объединения столбцов crash_date и crash_time
df['full_datatime'] = df['crash_date'].astype('str') + 'T' + df['crash_time'].astype('str')
```

```python
#меняем тип столбца datatime
df['full_datatime'] = pd.to_datetime(df['full_datatime'], format='%Y/%m/%dT%H:%M:%S')
df
```

```python
df.info()
```

```python
#из столбца full_datatime находим день недели
df['wd'] = df['full_datatime'].dt.weekday
```

```python
week = {0:'понедельник',
       1:'вторник',
       2:'среда',
       3:'четверг',
       4:'пятница',
       5:'суббота',
       6:'воскресенье'}
```

```python
#с помощью map сопоставляем числовое значение дня недели с его названием на русском языке
df['wd'] = df['wd'].map(week)
df.head()
```
