```python
import pandas as pd
```

```python
df = pd.read_excel("Алгоритм_Джонсона_данные.xlsx")
df
```

```python
def johnson_algorithm_dataframe(processing_times_df):
    df = processing_times_df.copy()
    min_a_jobs = df[df['Подготовка'] <= df['Выполнение']].copy()
    min_b_jobs = df[df['Подготовка'] > df['Выполнение']].copy() 

    min_a_jobs = min_a_jobs.sort_values(by='Подготовка')
    min_b_jobs = min_b_jobs.sort_values(by='Выполнение', ascending=False)

    optimal_order_df = pd.concat([min_a_jobs, min_b_jobs])
    optimal_order = optimal_order_df['Задача'].values

    return optimal_order
```

```python
johnson_algorithm_dataframe(df)
```
