Какие основные тренды в продаже объектов наблюдаются в разрезе спроса? (определите правила вымываемости)


```python
demand['Год'] = demand['Дата\nдоговора'].dt.year

demand_trends = demand.groupby(['Тип\nплощади', 'Форма оплаты', 'Год'])['Стоимость'].count().reset_index()
demand_trends = demand_trends.rename(columns={'Стоимость': 'Количество сделок'})

plt.figure(figsize=(14, 7))
sns.barplot(data=demand_trends, x='Тип\nплощади', y='Количество сделок', hue='Форма оплаты', palette='Set1')
plt.title('Тренды спроса по типам квартир и формам оплаты (по годам)', fontsize=14)
plt.xlabel('Тип площади', fontsize=12)
plt.ylabel('Количество сделок', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Форма оплаты', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()
```

```python
demand['Год'] = demand['Дата\nдоговора'].dt.year
demand['Месяц'] = demand['Дата\nдоговора'].dt.month

payment_trends = demand.groupby(['Год', 'Месяц', 'Форма оплаты'])['Стоимость'].count().reset_index()
payment_trends = payment_trends.rename(columns={'Стоимость': 'Количество сделок'})

total_deals_per_month = payment_trends.groupby(['Год', 'Месяц'])['Количество сделок'].sum().reset_index()
total_deals_per_month = total_deals_per_month.rename(columns={'Количество сделок': 'Общее количество сделок'})

payment_trends = pd.merge(payment_trends, total_deals_per_month, on=['Год', 'Месяц'])

payment_trends['Процент'] = payment_trends['Количество сделок'] / payment_trends['Общее количество сделок'] * 100

payment_pivot = payment_trends.pivot_table(index=['Год', 'Месяц'], columns='Форма оплаты', values='Процент', fill_value=0)

plt.figure(figsize=(16, 8))
payment_pivot.plot(kind='bar', stacked=True, ax=plt.gca(), colormap='viridis')
plt.title('Доли форм оплаты по месяцам ', fontsize=16)
plt.xlabel('Год - Месяц', fontsize=14)
plt.ylabel('Доля (%)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Форма оплаты', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.grid(axis='y')
plt.show()
```

```python
type_app_count_2 = demand.pivot_table(
    index=[demand['Дата\nдоговора'].dt.to_period("M"),'Проект', 'Тип\nплощади'],
    values='СВЦ',
    aggfunc=['count','mean']
).reset_index()
type_app_count_2.columns = ['Дата_договора','Проект', 'Тип_площади', 'Количество', 'Средний_СВЦ']

type_app_count_2['Дата_договора'] = type_app_count_2['Дата_договора'].dt.to_timestamp()

for project in type_app_count_2['Проект'].unique():
    project_data = type_app_count_2[type_app_count_2['Проект'] == project]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle(f'Динамика продаж и СВЦ в проекте "{project}"')


    num_area_types = len(project_data['Тип_площади'].unique())
    width = 0.8 / num_area_types
    x = project_data['Дата_договора'].unique()


    for i, area_type in enumerate(project_data['Тип_площади'].unique()):
        area_type_data = project_data[project_data['Тип_площади'] == area_type]


        common_index = pd.Index(x)
        area_type_data = area_type_data.set_index('Дата_договора').reindex(common_index).fillna(0)

        x_offset_days = (i - (num_area_types - 1) / 2) * width * 30
        x_positions = area_type_data.index + pd.Timedelta(days=x_offset_days)

        ax1.bar(x_positions, area_type_data['Количество'],
                width=pd.Timedelta(days=width * 30), label=area_type, alpha=0.7)

    ax1.set_xlabel('Дата')
    ax1.set_ylabel('Число реализованных квартир')
    ax1.set_title('Динамика продаж')
    ax1.grid(True)

    ax1.xaxis.set_major_locator(mdates.MonthLocator())
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.setp(ax1.get_xticklabels(), rotation=45, ha="right")

    ax1.legend(title='Тип площади')

    for area_type in project_data['Тип_площади'].unique():
        area_type_data = project_data[project_data['Тип_площади'] == area_type]
        ax2.plot(area_type_data['Дата_договора'], area_type_data['Средний_СВЦ'], marker='o', linestyle='-', label=area_type)

    ax2.set_xlabel('Дата')
    ax2.set_ylabel('Средний СВЦ')
    ax2.set_title('Динамика СВЦ')
    ax2.grid(True)
    ax2.tick_params(axis='x', rotation=45, labelrotation=45)
    ax2.legend(title='Тип площади')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
